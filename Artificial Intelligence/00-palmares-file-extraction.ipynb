```python
# Open the PDF file
Myoptions = []
MyProvinces = []
with open('Palmares_KINSHASA-EST_11_2015.pdf', 'rb') as file:
    reader = PdfReader(file) 
    num_pages = len(reader.pages)
    for page_num in range(num_pages):
        page = reader.pages[page_num]
        text = page.extract_text() 


        # Regular expressions to match options and provinces
        option_pattern = r"Option\s+:\s+([^\-]+)"
        province_pattern = r"Province\s+:\s+([^\-]+)"

        # Find all matches in the text
        options = re.findall(option_pattern, text)
        provinces = re.findall(province_pattern, text)

        # Remove duplicates by converting to a set and back to a list
        unique_options = list(set(options))
        unique_provinces = list(set(provinces))
        # Cleaning the results by stripping extra whitespaces
        unique_options = [opt.strip() for opt in unique_options]
        unique_provinces = [prov.strip() for prov in unique_provinces]
        
        for opt in unique_options:
            Myoptions.append(opt)
        for pro in unique_provinces:
            MyProvinces.append(pro)

Myoptions = list(set(Myoptions))
MyProvinces = list(set(MyProvinces))        

print("Unique Options:", Myoptions)
print("Unique Provinces:", MyProvinces)
```
